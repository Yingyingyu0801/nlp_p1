```python
def uni_generator(name, k=1, unkp=0.0005):    
    ps = PorterStemmer()

    # count word frequency
    count = defaultdict(int)
    f = open("./train/" + name + ".txt", "r")
    reviews = []
    for review in f:
        reviews.append(review)
        words = review.split(" ")
        
        # preprocess

        
        for word in words:
            word = ps.stem(word)
            count[word] += 1
            
    words = []
    for word in count:
        if count[word] == 1:
            words.append(word)
    return words
```
